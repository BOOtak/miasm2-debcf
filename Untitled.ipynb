```python
from miasm2.analysis.machine import Machine
from miasm2.ir.ir import AssignBlock, IRBlock
from miasm2.analysis.data_flow import dead_simp, remove_empty_assignblks, \
    merge_blocks

start_addr = offset

ir_arch = ira(mdis.loc_db)

entry_points = set([start_addr])
ircfg = ir_arch.new_ircfg_from_asmcfg(disasm)

simplified = True
if simplified:
    dead_simp(ir_arch, ircfg)
    ircfg.simplify(expr_simp)
    modified = True
    while modified:
        modified = False
        modified |= dead_simp(ir_arch, ircfg)
        modified |= remove_empty_assignblks(ircfg)
#         modified |= merge_blocks(ircfg, entry_points)
```

```python
prologue = None
dispatcher = None
predispatcher = None

relevant = set()

for node in ircfg.nodes():
    if len(ircfg.predecessors(node)) == 0:
        prologue = node
        # prologue is relevant
        relevant.add(node)
        break

for node in ircfg.nodes():
    if prologue in ircfg.predecessors(node):
        dispatcher = node
        break

for node in ircfg.nodes():
    if len(ircfg.successors(node)) == 1 and ircfg.successors(node)[0] == dispatcher and node is not prologue:
        predispatcher = node
        break
        
# function-specific!
# dispatcher = list(ircfg.getby_offset(0x431))[0].loc_key
# predispatcher = list(ircfg.getby_offset(0x713))[0].loc_key
# prologue = list(ircfg.getby_offset(0x41a))[0].loc_key
# relevant.add(list(ircfg.getby_offset(0x400))[0].loc_key)
# relevant.add(list(ircfg.getby_offset(0x41a))[0].loc_key)

# preprologue = set([list(ircfg.getby_offset(0x400))[0].loc_key])

# for node in ircfg.nodes():
#     if node.key == 51 or node.key == 52:
#         relevant.add(node)
#         preprologue.add(node)
# /function-specific!

print("prologue: %s" %prologue)
print("predispatcher: %s" % predispatcher)
print("dispatcher: %s" % dispatcher)
print
        
for node in ircfg.nodes():
    if len(ircfg.successors(node)) == 1 and ircfg.successors(node)[0] == predispatcher:
        # it may be unconditional original basic block, or fall-through of conditional one
        if len(ircfg.predecessors(node)) == 1:
            # it is unconditional original basic block
            relevant.add(node)
        elif len(ircfg.predecessors(node)) == 2:
            # it is fall-through of conditional relevant block, let's find it!
            parents = ircfg.predecessors(node)
            grandparents = []
            for p in parents:
                grandparents.extend(ircfg.predecessors(p))
            for p in parents:
                if p in grandparents:
                    relevant.add(p)
                    break
        else:
            print("Not sure what to do with node leading to predispatcher having %s parents" % len(ircfg.predecessors(node)))
    elif len(ircfg.successors(node)) == 0:
        # return is relevant
        relevant.add(node)

print("Relevant basic blocks: %s" % len(relevant))
print("="*24)
for r in relevant:
    offset = ircfg.loc_db.get_location_offset(r)
    print(hex(offset) if offset is not None else r)
```

```python
# Init our symbols with all architecture known registers
symbols_init =  {}
for i, r in enumerate(all_regs_ids):
    symbols_init[r] = all_regs_ids_init[i]

# Create symbolic execution engine
symb = symbexec(ir_arch, symbols_init)
# print("[ESP - 0x18] = %s" % symb.symbols.read(ExprMem(ExprId('ESP_init', 32)+ExprInt(0xFFFFFFE4, 32), 32)))
# print("EAX = %s" % symb.symbols.read(ExprId('EAX', 32)))

# list of unvisited branches with corresponding states
unvisited = []

current_branch_kind = True

# Get the block we want and emulate it
# We obtain the address of the next block to execute
block = ircfg.get_block(list(ircfg.nodes())[0])
relevant_parent = prologue

# Control flow dictionnary {parent: set({child: branch_kind})
flow = {}

# Init flow dictionnary with empty sets of childs
for r in relevant: flow[r] = []

while True:
    if block.loc_key in relevant and block.loc_key != prologue and block.loc_key not in preprologue:
        flow[relevant_parent].append({block.loc_key: current_branch_kind})
        relevant_parent = block.loc_key

    nxt_addr = symb.eval_updt_irblock(block, step=False)

    # Run the Miasm's simplification engine on the next
    # address to be sure to have the simplest expression
    simp_addr = expr_simp(nxt_addr)
    
    # The simp_addr variable is an integer expression (next basic block offset)
    if isinstance(simp_addr, ExprInt):
#         print("Jump on next basic block: %s" % simp_addr)
        block = ircfg.get_block(simp_addr)
    # The simp_addr variable is a condition expression
    elif isinstance(simp_addr, ExprCond):
        branch1,branch2 = simp_addr.src1, simp_addr.src2 # src1 = true, src2 = false
        print("Condition: %s? %s or %s" % (simp_addr.cond, branch1,branch2))
        block = ircfg.get_block(branch1)
        # traverse "False" branch now, push "True" branch for later use
        current_branch_kind = True
        unvisited.append((branch2, symb.state, relevant_parent))
    elif isinstance(simp_addr, ExprMem):
        print("Got in Return statement!: %s" % str(simp_addr))
        if len(unvisited) > 0:
            offset, st, relevant_parent = unvisited.pop()
#             print("Go to branch at %s instead" % offset)
            symb.state = st
            current_branch_kind = False
            block = ircfg.get_block(offset)
        else:
            print("no more nodes to visit")
            break
    else:
        print("Unsupported statement!: %s (%s)" % (str(simp_addr), str(type(simp_addr))))
        block = ircfg.get_block(simp_addr)
        continue
```

```python
for i, v in flow.iteritems():
    offset = ircfg.loc_db.get_location_offset(i)
    children = []
    for child in v:
        for chloc, kind in child.iteritems():
            choff = ircfg.loc_db.get_location_offset(chloc)
            children.append("%s:%s, " % (hex(choff) if choff is not None else "[VIRTUAL]", kind))
    print("%s -> %s" % (hex(offset) if offset is not None else i, children))
```

```python
fixed = ir_arch.new_ircfg()

def off(block):
    loc = block.loc_key
    offset = ircfg.loc_db.get_location_offset(loc)
    return hex(offset) if offset is not None else str(loc)

def get_last_expr(block):
    ablocks = block.assignblks
    last_ablock = ablocks[-1]
    return last_ablock.values()[-1]

def my_mod_src(src, chlok, kind, last_expr):
    if src == last_expr:
        if isinstance(last_expr, ExprCond):
            cond = src.cond
            branch1, branch2 = src.src1, src.src2
            if kind:
                branch1 = ExprLoc(chloc, branch1.size)
            else:
                branch2 = ExprLoc(chloc, branch2.size)
            return ExprCond(cond, branch1, branch2)
        elif isinstance(last_expr, ExprLoc):
            return ExprLoc(chloc, src.size)
    else:
        return src

for r in relevant:
    parent_block = ircfg.get_block(r)
    block_in_flow = flow[r]
    last_expr = get_last_expr(parent_block)
    if len(block_in_flow) > 0:
        for child in block_in_flow:
            for chloc, kind in child.iteritems():
                if fixed.get_block(parent_block.loc_key) is not None:
                    parent_block = fixed.get_block(parent_block.loc_key)
                    last_expr = get_last_expr(parent_block)
                new_parent = parent_block.modify_exprs(None, lambda x : my_mod_src(x, chloc, kind, last_expr))
                if fixed.get_block(parent_block.loc_key) is not None:
                    fixed.del_node(parent_block.loc_key)
                fixed.add_irblock(new_parent)
    else:
        fixed.add_irblock(parent_block)

# add edges that wwere not automatically added while building fixed CFG
for node in fixed.nodes():
    if node in flow.keys():
        block_in_flow = flow[node]
        if len(block_in_flow) > 0:
            for child in block_in_flow:
                for chloc, kind in child.iteritems():
                    fixed.add_uniq_edge(node, chloc)

# simplify CFG
dead_simp(ir_arch, fixed)
ircfg.simplify(expr_simp)
modified = True
while modified:
    modified = False
    modified |= dead_simp(ir_arch, fixed)
    modified |= remove_empty_assignblks(fixed)
    modified |= merge_blocks(fixed, entry_points)

dead_nodes = []
# remove dead nodes
for node in fixed.nodes():
    if len(fixed.successors(node)) == 0 and len(fixed.predecessors(node)) == 0:
        dead_nodes.append(node)

for node in dead_nodes:
    fixed.del_node(node)
```
